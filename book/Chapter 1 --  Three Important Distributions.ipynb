```julia
#semicircle.jl
#Algorithm 1.2 of Random Eigenvalues by Alan Edelman

#Experiment:  Sample random symmetric Gaussian matrices
#Plot:        Histogram of the eigenvalues
#Theory:      Semicircle as n->infinity


## Parameters
n = 1000
t = 1


## Experiment
vals = Float64[]                            # initialize result vector
for i = 1:t
    A = randn(n, n)                         # draw n by n matrix with idependent gaussians
    S = (A + A')/2                          # symmetrize matrix
    vals = append!(vals, eigvals(S))        # calculate eigenvalues and append to result vector
end





vals /= âˆš(n/2)                              # scale eigenvalues

xlims!(-3, 3) # set limits for x axis in plot
gr()
histogram(vals, nbins=50, normed = true, color="yellow", label="")                 # make histogram
xvals = linspace(-2, 2)
plot!(xvals, sqrt.(4 - xvals.^2)/(2*Ï€),color="blue",linewidth = 2.0, title="Semi-Circle Law", label="") # plot semi circle
```

```julia
#tracywidom.jl
#Algorithm 1.3 of Random Eigenvalues by Alan Edelman

#Theory:  Compute and plot the Tracy-Widom distribution

##Parameters
t0 =  5.0;   # right endpoint
tn = -8.0;   # left endpoint
dx = .005; # discretization

##Theory: The differential equation solver
function TracyWidom(du,u,p,t)
 du[1] = u[2]
 du[2] = t*u[1] + 2*u[1]^3
 du[3] = u[4]
 du[4] = u[1]^2
end

u0 = [airyai(t0);airyaiprime(t0);0.0;airyai(t0)^2]
tspan = (t0,tn)
prob = ODEProblem(TracyWidom,u0,tspan);
sol = solve(prob, Vern8(),reltol=1e-12,abstol=1e-12);

F2 = exp.(-sol[3,:])
f2 = gradient(F2,sol.t)

## Plot
x = linspace(-2,2)
plot(sol.t,f2,xlims=(-5,2))
```
